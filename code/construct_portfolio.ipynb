```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib as mpl
mpl.style.use('seaborn')
```

```python
path = 'C:/Users/dell/Desktop/semester2/statistics/project/data.csv'
Y = pd.read_csv(path)
Y = Y.drop('Unnamed: 0', 1)
path = 'C:/Users/dell/Desktop/semester2/statistics/project/state.csv'
X = pd.read_csv(path)
X = X.drop('Unnamed: 0', 1)
Y = np.array(Y)
Y = np.transpose(Y)
X = np.array(X)
X = np.transpose(X)
len(X)
```

```python
#for AAA
whole_weight = [0.6566,0.8255,0.904,0.8826,0.8127,0.5872,0.323,-0.0454,-0.4846,-0.8183]
normal_weight = [1.11891,1.1745,1.20190,  1.10228,  0.94948,  0.67601,  0.31610, -0.09458, -0.54190, -0.94059]
crisis_weight = [0.527181,  0.703220,  0.782795,  0.778639,  0.735349,  0.531349,  0.306626, -0.038049, -0.467345,-0.771974]
whole = np.dot(np.array(whole_weight),X[1:])
crisis = np.dot(np.array(crisis_weight),X[1:])
normal = np.dot(np.array(normal_weight),X[1:])

r = [1]
ben_r = [1]
ben_r.extend(np.cumprod([1+(Y[0][i]-whole[i])/(12*(sum(whole_weight) - 1)) for i in range(215)]))

for i in range(215):
    if X[0][i] == 'State 2':
        r.append(r[-1]*(1+(Y[0][i]-crisis[i])/(12*(sum(crisis_weight) - 1))))
    else:
        r.append(r[-1]*(1+(Y[0][i]-normal[i])/(12*(sum(normal_weight) - 1))))

plt.figure(figsize=(20,5))

plt.subplot(131)
plt.plot(ben_r, label = 'Whole Sample')
plt.plot(r, label = 'Regime Switching Model')
plt.title('Asset')
plt.legend()

plt.subplot(132)
s = [(r[i+1]-r[i])/r[i] for i in range(215)]
plt.plot(s)
plt.title('Return')

plt.subplot(133)
plt.plot([(r[i+1]-r[i])/r[i] - (ben_r[i+1]-ben_r[i])/ben_r[i] for i in range(215)])
plt.title('r_{regime_switching} - r_{whole sample}')

plt.show()
print('Return = ', np.mean(s)*12*100,'%')
print('Sharpe Ratio = ',np.mean(s)/np.std(s))
```

```python
#for AAA
x = np.arange(10)
plt.bar(x-0.2, normal_weight,width=0.2,color='b',align='center',label = 'normal')
plt.bar(x, whole_weight,width=0.2,color='g',align='center', label = 'whole sample')
plt.bar(x+0.2, crisis_weight,width=0.2,color='r',align='center', label = 'crisis')
plt.title('Weights in Different States for AAA')
plt.legend()
plt.show()
```
